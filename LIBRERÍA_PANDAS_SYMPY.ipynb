```python
import miceforest as mf ##cargamos la libreria miceforest
from sklearn.datasets import load_iris ##Estraigo el dataset iris
import numpy as np ## Importamos numpy, es para cálculos numéricos
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
df_iris = pd.DataFrame(data=iris.data,columns = iris.feature_names)
df_iris['target'] = iris.target
df_iris
```

```python
df_iris.columns
```

```python
## Renombrar la columna target con species
df_iris.rename({"target":"species"}, inplace = True, axis=1)
df_iris
```

```python
##transformo el tipo de dato del campo species
df_iris['species'] = df_iris['species'].astype('category')
df_iris.info()
```

```python
## Este dataset tiene todos sus valores no-nulos, solo queresmos asignar algunos nulos para efectos del ejercicio
## Asigno valores nulos a la columna de especies
## perce = 0.25 significa que voy a asignar valores nulos al 25% de los datos, eso se hace de formal aleatoria, no se cuales datos quedan con NaN
iris_amp = mf.ampute_data(df_iris, perc = 0.25, random_state = 1991) #Realizo una amputación de datos aplicando NaN
iris_amp
```

```python
##Crear Kernel
kds = mf.ImputationKernel(
    iris_amp,                 ##NOmbre del dataframe para entrenar modelo
    random_state =1991        ##Valores aleatorios para entrenar modelo
)

#Correr el algoritmo MICE dutante 2 iteracciones
kds.mice(2)

# Retorna el dataset completo.
iris_completo = kds.complete_data()
iris_completo
```
